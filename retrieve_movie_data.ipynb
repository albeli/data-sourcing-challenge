### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
from urllib.parse import quote
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python

# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"


# filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'
filter_query = 'section_name:Movies AND type_of_material:Review AND headline:love'

encoded_filter_query = quote(filter_query)

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
# query_url = f"{url}api-key={nyt_api_key}&q={filter_query}&fq={filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}"

query_url = f"{url}api-key={nyt_api_key}&fq={encoded_filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}"

query_url
```

```python
# Create an empty list to store the reviews
reviews_list = []



# loop through pages 0-19
for page in range(0, 20):

    # create query with a page number
    # API results show 10 articles at a time
    query_url_page = f"{query_url}&page={page}"


    
    # Make a "GET" request and retrieve the JSON
    response = requests.get(query_url_page).json()
    reviews = response


    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)

    
    # Try and save the reviews to the reviews_list
    try:
        

        # loop through the reviews["response"]["docs"] and append each review to the list
        for review in reviews["response"]["docs"]:
            reviews_list.append(review)

        # Print the page that was just retrieved
        print(f"Checked page {page}")



        # Print the page number that had no results then break from the loop
    except:
        print(f"Page {page} does not have results")
        break
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
print(json.dumps(reviews_list[0:5], indent=4))

```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
reviews_df = pd.json_normalize(reviews_list)
reviews_df
```

```python

# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

reviews_df['title'] = reviews_df['headline.main'].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])


reviews_df


```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    if isinstance(keyword_list, list):
        for item in keyword_list:
            if isinstance(item, dict) and 'name' in item and 'value' in item:
                keyword = f"{item['name']}: {item['value']};"
                extracted_keywords += keyword
    return extracted_keywords

reviews_df['keywords'] = reviews_df['keywords'].apply(extract_keywords)
reviews_df
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
titles = reviews_df["title"].to_list()
titles

```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
import requests
import time

# Initialize variables
tmdb_movies_list = []
request_counter = 1
search_url = "https://api.themoviedb.org/3/search/movie?query="

# Loop through the titles list
for title in titles:
    # Construct search query
    query_url = search_url + title + tmdb_key_string

    # Make search request
    search_response = requests.get(query_url).json()

    # Rate limiting
    if request_counter % 50 == 0:
        print("Sleeping...")
        time.sleep(1)
    request_counter += 1

    try:
        # Extract movie ID from first result
        movie_id = search_response['results'][0]['id']

        # Make movie detail request
        movie_url = f"https://api.themoviedb.org/3/movie/{movie_id}?{tmdb_key_string}"
        movie_response = requests.get(movie_url).json()

        # Extract details
        genres = [genre['name'] for genre in movie_response['genres']]
        spoken_languages = [lang['english_name'] for lang in movie_response['spoken_languages']]
        production_countries = [country['name'] for country in movie_response['production_countries']]

        # Create a dictionary of movie details
        movie_dict = {
            'title': title,
            'original_title': movie_response['original_title'],
            # Add other required fields here
            'genres': genres,
            'spoken_languages': spoken_languages,
            'production_countries': production_countries
        }

        # Append to the list
        tmdb_movies_list.append(movie_dict)
        print(f"Movie found and processed: {title}")

    except (IndexError, KeyError):
        print(f"Movie not found: {title}")
```

```python
# Create an empty list to store the results
tmdb_movies_list = []


# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 0



# Loop through the titles
for title in titles:
    # Check if we need to sleep before making a request
    if request_counter % 50 == 0 and request_counter != 0:
        print("Sleeping for 1 seconds")
        time.sleep(1)


    # Add 1 to the request counter
    request_counter += 1

    
    # Perform a "GET" request for The Movie Database
    response = requests.get(url + title.replace(" ", "%20") + tmdb_key_string).json()
    movie = response


    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    try:

        # Get movie id
        movie_id = movie["results"][0]["id"]

        

        # Make a request for a the full movie details
        movie_url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_api_key}"



        # Execute "GET" request with url
        movie_response = requests.get(movie_url).json()
        

        
        # Extract the genre names into a list           
        genres = [genre["name"] for genre in movie_response["genres"]]


        # Extract the spoken_languages' English name into a list
        spoken_languages = [language["english_name"] for language in movie_response["spoken_languages"]]


        # Extract the production_countries' name into a list
        production_countries = [country["name"] for country in movie_response["production_countries"]]


        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        tmdb_movies_list.append({
            "title": movie_response["title"],
            "original_title": movie_response["original_title"],
            "budget": str(movie_response["budget"]),
            "original_language": movie_response["original_language"],
            "homepage": movie_response["homepage"],
            "overview": movie_response["overview"],
            "popularity": movie_response["popularity"],
            "runtime": movie_response["runtime"],
            "revenue": str(movie_response["revenue"]),
            "release_date": movie_response["release_date"],
            "vote_average": movie_response["vote_average"],
            "vote_count": movie_response["vote_count"],
            "genres": genres,
            "spoken_languages": spoken_languages,
            "production_countries": production_countries
        })

        
        # Print out the title that was found
        print(f"Found: {title}")

    except:
            # Print out the title that was not found
            print(f"Movie not found: {title}")

```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
print(json.dumps(tmdb_movies_list[0:5], indent=4))
```

```python
# Convert the results to a DataFrame
tmdb_movies_df = pd.DataFrame(tmdb_movies_list)
tmdb_movies_df.head()
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews_df and TMDB tmdb_movies_df on title
merged_df = reviews_df.merge(tmdb_movies_df, on='title')

```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ['genres', 'spoken_languages', 'production_countries']



# Create a list of characters to remove
characters_to_remove = ['[', ']', "'"]

# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    merged_df[column] = merged_df[column].astype(str) 
    # Loop through characters to remove
    for char in characters_to_remove:
        merged_df[column] = merged_df[column].str.replace(char, '')
    

    


# Display the fixed DataFrame

merged_df.head()




```

```python
# Drop "byline.person" column
merged_df.drop(columns="byline.person", inplace=True)

```

```python
# Delete duplicate rows and reset index
merged_df.drop_duplicates(inplace=True)
merged_df.reset_index(drop=True, inplace=True)
merged_df.head()
```

```python
# Export data to CSV without the index
merged_df.to_csv('Output/merged_data.csv', index=False)
```
